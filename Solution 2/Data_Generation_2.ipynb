```python
import pandas as pd
import math
import random

m=[]
c1=[]
c2=[]

for i in range(0,1001):
    m.append(round(random.uniform(-10,10),2))
    c1.append(round(random.uniform(-10,10),2))
    c2.append(round(random.uniform(-10,10),2))
    
Distance=[]
for i in range (0,1001):
    Distance.append(abs(c1[i]-c2[i]))


Data = pd.DataFrame()
Data['m'] = m
Data['c1'] = c1
Data['c2'] = c2
Data['Distance'] = Distance

Data.to_csv('Distance_file_2.csv',index=False)
```
